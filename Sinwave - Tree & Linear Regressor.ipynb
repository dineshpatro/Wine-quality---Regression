```python
from __future__ import print_function
import sklearn
from IPython.display import display, HTML
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
os.chdir(r'C:\Users\Dinesh\Documents\Elite Data Science\Project 1 Workbook Bundle\project_files')
```

```python
# 100 evenly spaced numbers between 0 and 2*np.pi
x = np.linspace(0, 2*np.pi, 100)
np.random.seed(321)
noise = np.random.normal(0,0.5,100)

y = np.sin(x) + noise
```

```python
df = pd.DataFrame({'x': x,'y':y})
```

```python
plt.scatter(df.x, df.y)
plt.plot(df.x, np.sin(df.x), color = 'c')
plt.show()
```

```python
#mean model
pred = np.mean(df.y)
print(pred)
```

```python
#black line representing the model
plt.scatter(df.x, df.y)
plt.plot(df.x, [pred]*len(df.x), 'k--')
plt.show()
```

```python
lm = LinearRegression()
features = df.drop('y', axis = 1)
target = df.y
```

```python
lm.fit(features, target)
```

```python
print (lm.intercept_)
print (lm.coef_)
```

```python
plt.scatter(df.x, df.y)
plt.plot(df.x, lm.predict(features), 'k--')
plt.show()
```

```python
def fitandPlot(df, model):
    features = df.drop('y', axis = 1)
    target = df.y
    model.fit(features, target)
    plt.scatter(df.x, df.y)
    plt.plot(df.x, model.predict(features), 'k--')
    plt.show()
```

```python
df['x_squared'] = df.x**2
```

```python
df['x_cubed'] = df.x**3
```

```python
df['x_4'] = df.x**4
```

```python
lm = LinearRegression()
fitandPlot(df, lm)
```

```python
fitandPlot(df, DecisionTreeRegressor())
```
